```python
from qdrant_client import QdrantClient

client = QdrantClient(url="http://localhost:6333")   # or cloud URL / API key

def all_points(collection: str, batch: int = 1000,
               *, with_vectors=False, with_payload=True):
    """Yield every point in <collection>, batch-reading <batch> at a time."""
    offset = None
    while True:
        points, offset = client.scroll(
            collection_name=collection,
            limit=batch,
            offset=offset,
            with_vectors=with_vectors,
            with_payload=with_payload,
        )
        for p in points:
            yield p
        if offset is None:           # no more pages
            break

for rec in all_points("sn_rulings", with_vectors=False):
    print(rec.id, rec.payload)
```

```python
from itertools import groupby

all_records = []
all_jsonl_bytes = []
with open('batch_extract_out.jsonl', "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            all_records.append(json.loads(line))
            # print(all_records[-1])
        else:
            all_records.append(None)
all_records = [json.loads(r['response']['body']['output'][1]['content'][0]['text']) for r in all_records]  
print(all_records[0])
for r in all_records:
    merged = []
    for k, group in groupby(r['paragraphs'], key=lambda d: d['section']):
        grp = list(group)
        # Single element → keep as–is, otherwise merge
        merged.append({**grp[0], 'para_no': grp[0]['para_no'] if len(merged) == 0 else merged[-1]['para_no']+1}
                      if len(grp) == 1
                      else {'section': grp[0]['section'],
                                                    'para_no': grp[0]['para_no'] if len(merged) == 0 else merged[-1]['para_no']+1,
                                                    'text': grp[0]['text']+grp[1]['text']})
    all_jsonl_bytes.append({'paragraphs': merged})
# for m in all_jsonl_bytes:
#     for p in m['paragraphs']:
#         print(p['para_no'])
with open('batch_input_enriched.jsonl', 'w') as f:
    f.write("\n".join([json.dumps(r) for r in all_jsonl_bytes]))
```
